```python
from power import *
from power.systems import *
from opf_linear.opf_loss import LinearDispatch
from utils.load_scen import apply_load_scen
from utils.wnd_scen import apply_wnd_scen
import numpy as np
from opf_linear.utils.extr_and_save import extract_and_save_results
from pathlib import Path 
import copy
```

```python
# nets = [B6L8EOL(), IEEE118EOL()]
# nets = [B6L8EOL()]
# nets = [IEEE118EOL()]
nets = [B3()]
seed = 42
# rng = np.random.d2fault_rng(seed=seed)
rng = np.random.default_rng(seed=seed)
nscen = 2
```

```python
for net in nets:
    results_dir = Path(f"results_{net.name}")
    results_dir.mkdir(exist_ok=True)
    for scen in range(nscen):
        scenario_dir = results_dir / f"scenario_{scen}"
        scenario_dir.mkdir(parents=True, exist_ok=True) # Cria a pasta do cenário (e a principal, se necessário)
        apply_wnd_scen(net=net, rng=rng)
        apply_load_scen(net=net, rng=rng)

        # Rodar o problema sem CTG:
        solver = LinearDispatch(net)
        status, _, _ = solver.solve_loss()
        solver = LinearDispatch(net)
        status, _, _ = solver.solve_loss()
        if status == 'Optimal':
            output_filename = scenario_dir/f"cenario_{scen}_BASE_CASE.json"
            extract_and_save_results(solver.problem, solver.net, output_filename)
        else:
            print("DEU RUIM")

        for idx, line in enumerate(net.lines):
            ctg_net = copy.deepcopy(net)
            ctg_net.lines.pop(idx)
            print(ctg_net.lines)
            solver = LinearDispatch(ctg_net)
            status, _, _ = solver.solve_loss()
            if status == 'Optimal':
                output_filename = scenario_dir/f"cenario_{scen}_ctg{idx}.json"
                extract_and_save_results(solver.problem, solver.net, output_filename)
            else:
                print("DEU RUIM")

```
